## **Multiple linear regression - prediction heating and cooling loads**

###### This example tries to correlate heating and cooling loads of a building to its relative compactness, surface area, wall area and roof area

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics

import seaborn as sb


%matplotlib inline
```

#### **Importing Data**

```python
#ds = pd.read_csv(".\\data\\energy.csv")
ds=pd.read_csv('https://raw.githubusercontent.com/sekhargullapalli/exploring-data-science/master/data/energy.csv')
print(ds.shape)
ds.describe()
```

### **Exploratory Data Analysis**
##### **Pair Plots**

```python
#sb.pairplot(ds[['Relative Compactness','Surface Area','Wall Area','Roof Area','Orientation','Heating Load','Cooling Load']], hue='Orientation', height=3, aspect=2)
sb.pairplot(
            ds[['Relative Compactness','Surface Area','Wall Area','Roof Area','Heating Load','Cooling Load']],
            diag_kind='kde',
            plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
            height=2
            )
plt.suptitle('Pair Plot for Energy Data')
```

#### **PairGrid**

```python
# Function to calculate correlation coefficient between two arrays
def corr(x, y, **kwargs):    
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))    
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.2, 0.95), size = 10, xycoords = ax.transAxes)
```

```python
grid = sb.PairGrid(data= ds,
                    vars = ['Relative Compactness','Surface Area','Wall Area','Roof Area','Heating Load','Cooling Load'],
                    height = 2)
#scatter plots in the upper triangle
grid = grid.map_upper(plt.scatter, color = 'darkred')
grid = grid.map_upper(corr)

#historgram in the diagonal
grid = grid.map_diag(plt.hist, bins = 10, color = 'darkred', 
                     edgecolor = 'k')
#density plots in lower triangle
grid = grid.map_lower(sb.kdeplot, cmap = 'Reds')
```
